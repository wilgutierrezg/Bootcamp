1. Crear SparkContext

```python
from pyspark import SparkContext

sc = SparkContext("local", "DemoRDD")
```

2. Crear RDD desde colección

```python
datos = [
    ("ventas", 100),
    ("ventas", 200),
    ("marketing", 50),
    ("ventas", 150),
    ("marketing", 100)
]

rdd = sc.parallelize(datos)

print("RDD original:")
print(rdd.collect())
```

3. Transformaciones

```python
# Filtrar solo ventas
rdd_filtrado = rdd.filter(lambda x: x[0] == "ventas")

# Mapear solo valores
rdd_valores = rdd_filtrado.map(lambda x: (x[0], x[1] * 2))

# Agregar por clave
rdd_resultado = rdd_valores.reduceByKey(lambda a, b: a + b)

print("\nTransformaciones definidas (aún no ejecutadas)")
```

4. Acción

```python
print("\nResultado final:")
print(rdd_resultado.collect())
```

5. Ejemplo de acciones adicionales

```python
print("\nCantidad de registros:")
print(rdd_resultado.count())
```

6. Ver lineage (DAG)

```python
print("\nLineage del RDD:")
print(rdd_resultado.toDebugString())
```

7. Ejemplo de Lazy Evaluation

```python
rdd_lazy = rdd.map(lambda x: x[1] * 10)

print("\nLazy evaluation (no ejecuta aún)")

# Aquí recién ejecuta
print(rdd_lazy.collect())
```

8. Cache (opcional para mostrar performance)

```python
rdd_cache = rdd.filter(lambda x: x[1] > 50).cache()

print("\nPrimera ejecución (calcula):")
print(rdd_cache.collect())

print("\nSegunda ejecución (usa cache):")
print(rdd_cache.collect())
```

Fin

```python
sc.stop()
```
