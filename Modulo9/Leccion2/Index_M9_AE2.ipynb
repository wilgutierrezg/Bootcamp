**3) Pequeña prueba**

```python
data = [("Juan", 1200), ("Ana", 800), ("Pedro", 1500)]
df = spark.createDataFrame(data, ["cliente", "monto"])

df.show()
```

**4) mini demo de transformación + acción**

```python
df_filtrado = df.filter(df["monto"] > 1000)
df_resultado = df_filtrado.select("cliente", "monto")

df_resultado.show()
```

**Transformación**

```python
df_resultado = df.filter(df["monto"] > 1000).select("cliente", "monto")
```

```python
df_resultado
```

**Acción**

```python
df_resultado.count()
```

**Resumen:**

Partimos creando un entorno simple de trabajo en PySpark y construimos un DataFrame con algunos datos de ejemplo.

Luego aplicamos transformaciones como filter y select, con las que definimos qué queríamos hacer con los datos sin que Spark ejecutara nada todavía. Finalmente, al usar acciones como show() y count(), forzamos la ejecución del proceso completo, obteniendo primero los registros filtrados y luego el total de filas resultantes.

Con este flujo, vimos en la práctica cómo Spark separa la definición del trabajo de su ejecución, aplicando el concepto de lazy evaluation.

**Otro ejemplo con DataFrame en spark**

## Ahora veamos lo mismo… pero con la forma más básica de Spark

**RDD**

Un RDD es la forma más básica de trabajar datos en Spark

Es un conjunto de datos que está dividido en varias máquinas y que puedes procesar en paralelo



| Concepto  | Cómo pensarlo             |
| --------- | ------------------------- |
| DataFrame | tabla (tipo SQL / Pandas) |
| RDD       | lista distribuida         |


* RDD es la estructura base de Spark, pero en el día a día trabajamos más con DataFrames
- Un RDD es como una lista… pero distribuida en varias máquinas

```python
rdd = spark.sparkContext.parallelize([1, 2, 3, 4, 5])

rdd_doble = rdd.map(lambda x: x * 2)

rdd_filtrado = rdd_doble.filter(lambda x: x > 5)
```

```python
rdd_filtrado.collect()
```

## Ejemplo de código PySpark

```python
from pyspark.sql import SparkSession

# Iniciar una sesión de Spark
spark = SparkSession.builder \
    .appName("Ejemplo PySpark") \
    .getOrCreate()

# Crear un DataFrame a partir de un archivo CSV
df = spark.read.csv("datos.csv", header=True, inferSchema=True)

# Realizar transformaciones
resultado = df.filter(df.edad > 25) \
    .groupBy("departamento") \
    .count()

# Mostrar resultados
resultado.show()

# Cerrar la sesión
spark.stop()
```
