```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("PruebaColabPySpark") \
    .getOrCreate()

print(spark)
```

**Probar volumen “mediano” generado en el mismo notebook**

```python
n = 5_000_000

df = spark.range(0, n)
print(df.count())
df.show(5)
```

**Medir una operación más pesada**

```python
from pyspark.sql.functions import col

resultado = (
    df.withColumn("grupo", col("id") % 10)
      .groupBy("grupo")
      .count()
)

resultado.show()
```

```python
import time

start = time.time()
print(df.count())
print("Tiempo count:", round(time.time() - start, 2), "segundos")
```
