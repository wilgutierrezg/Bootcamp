1. Gráficos de dispersión (scatter plot)

- Relación entre edad y tarifa pagada, coloreando por supervivencia.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

titanic = sns.load_dataset("titanic")

plt.figure(figsize=(7,5))
sns.scatterplot(
    data=titanic,
    x="age",
    y="fare",
    hue="survived",
    alpha=0.7
)

plt.title("Relación entre Edad y Tarifa según Supervivencia")
plt.xlabel("Edad")
plt.ylabel("Tarifa (Fare)")
plt.show()
```

2. Matriz de correlación (heatmap)
- Relación lineal entre variables numéricas.

```python
import numpy as np

numeric_cols = titanic.select_dtypes(include=np.number)

corr = numeric_cols.corr()

plt.figure(figsize=(7,5))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Matriz de correlación - Titanic")
plt.show()
```

3. Gráficos de pares (pairplot)

```python
sns.pairplot(
    titanic,
    vars=["age", "fare", "sibsp", "parch"],
    hue="survived",
    corner=True
)
plt.show()
```

4. Mapas de calor (otra forma, más conceptual)

```python
tabla = pd.crosstab(titanic["pclass"], titanic["survived"], normalize="index")

plt.figure(figsize=(6,4))
sns.heatmap(tabla, annot=True, cmap="Blues", fmt=".2f")
plt.title("Supervivencia por Clase")
plt.xlabel("Sobrevive")
plt.ylabel("Clase")
plt.show()
```
