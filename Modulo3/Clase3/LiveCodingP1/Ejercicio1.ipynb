<h5 style="color:#1f77b4;">1. Importa la librería pandas</h5>

```python
import pandas as pd
```

<h5 style="color:#1f77b4;">2. Usa read_csv('pedidos.csv', sep=',') para cargar el CSV.</h5>

```python
# 2) Cargar el CSV con punto y coma
df_pedidos = pd.read_csv("pedidos.csv", sep=",")
print("CSV cargado:", df_pedidos.shape)
df_pedidos.head()
```

<h5 style="color:#1f77b4;">3. Usa read_excel('Clientes.xlsx', sheet_name=None) para cargar todas las hojas.</h5>

```python
dfs_excel = pd.read_excel("Clientes.xlsx", sheet_name=None)
print("Hojas en Excel:", list(dfs_excel.keys()))
```

```python
# Acceder a una hoja
df_clientes = dfs_excel["clientes"]
df_clientes.head()
```

```python
# Acceder a una hoja
df_empleados = dfs_excel["empleados"]
df_empleados.head()
```

<h5 style="color:#1f77b4;">4. Combina los datos relevantes en un único DataFrame.</h5>

```python
# mirar las columnas de todos los df
df_clientes.columns, df_empleados.columns, df_pedidos.columns
```

<p>1) Merge pedidos + clientes + empleados</p>


```python
# 1) Pedidos + Clientes (por IdCliente)
df_final = df_pedidos.merge(
    df_clientes,
    on="IdCliente",
    how="left"
)

# 2) + Empleados (por IdEmpleado)
df_final = df_final.merge(
    df_empleados,
    on="IdEmpleado",
    how="left"
)

df_final.head()
```

```python
df_final.columns
```

<p>Atención con la columna Cargo.</p>

```python
df_final = df_final.rename(columns={
    "Cargo_x": "CargoEnvio",
    "Cargo_y": "CargoEmpleado"
})
df_final.head()
```

<h5 style="color:#1f77b4;">5.Elimina columnas innecesarias con drop().</h5>

```python
df_final.columns
```

```python
# eliminamos: "Longitud","Latitud" 
df_final = df_final.drop(columns=["Longitud","Latitud"], errors="ignore")
```

```python
# finalmente nuestro df:
df_final.head()
```

<h5 style="color:#1f77b4;">6.Guarda con to_csv('consolidado.csv', index=False).</h5>

```python
# 6) Guardar consolidado a CSV
df_final.to_csv("consolidado.csv", index=False)
print("Archivo guardado")
```
