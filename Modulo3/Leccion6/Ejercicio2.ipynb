```python
import pandas as pd

# 0. Crear el set de datos para los departamentos
data = {
    'Departamento': ['Ventas', 'Ventas', 'TI', 'TI', 'RRHH', 'RRHH', 'Ventas', 'TI'],
    'Empleado': ['Ana', 'Luis', 'Carlos', 'Sofía', 'Elena', 'Pablo', 'Jorge', 'Marta'],
    'Puntuación': [85, 78, 92, 95, 60, 65, 88, 91]
}

df = pd.DataFrame(data)
df
```

```python
# --- PUNTO 1: Agrupar rendimiento por departamento ---
# Creamos el objeto de agrupación (esto no genera una tabla visual aún, sino un objeto interno)
agrupado_por_depto = df.groupby('Departamento')
print(agrupado_por_depto)
```

```python
# --- PUNTO 2: Calcular promedio, mínimo y máximo de puntuación ---
# Ahora aplicamos los cálculos sobre el objeto creado en el paso anterior
resumen_estadistico = agrupado_por_depto['Puntuación'].agg(['mean', 'min', 'max'])
resumen_estadistico
```

```python
# --- PUNTO 3: Filtrar resultados mayores a un umbral ---
umbral = 70
resumen_filtrado = resumen_estadistico[resumen_estadistico['mean'] > umbral]
resumen_filtrado
```

```python
# --- PUNTO 4: Ordenar de mayor a menor ---
resumen_ordenado = resumen_filtrado.sort_values(by='mean', ascending=False)
resumen_ordenado
```

```python
# --- PUNTO 5: Preparar reporte final para Excel ---
reporte_final = resumen_ordenado.rename(columns={
    'mean': 'Promedio',
    'min': 'Mínimo',
    'max': 'Máximo'
})

reporte_final
```

```python
# Exportar
reporte_final.to_excel('data/reporte_departamentos.xlsx')
```
