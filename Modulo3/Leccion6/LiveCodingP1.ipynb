<h3>PASO 0: Crear el set de datos</h3>

```python
import pandas as pd

data = {
    'Categoría': ['Electrónica', 'Electrónica', 'Hogar', 'Hogar', 'Electrónica', 'Hogar', 'Ropa', 'Ropa'],
    'Región': ['Norte', 'Sur', 'Norte', 'Sur', 'Norte', 'Norte', 'Sur', 'Norte'],
    'Ventas': [1200, 1500, 800, 1100, 1300, 950, 600, 750]
}

df = pd.DataFrame(data)
df
```

<h4>PASO 1 y 2: Agrupar por categoría y región + Calcular suma y promedio</h4>

```python
# Usamos .agg() para obtener ambos cálculos a la vez
resumen_estadistico = df.groupby(['Categoría', 'Región']).agg({
    'Ventas': ['sum', 'mean']
})

resumen_estadistico
```

<h4>PASO 3: Crear tabla dinámica para mostrar totales</h4>

```python

tabla_dinamica = df.pivot_table(
    values='Ventas', 
    index='Categoría', 
    columns='Región', 
    aggfunc='sum', 
    fill_value=0
)
tabla_dinamica
```
