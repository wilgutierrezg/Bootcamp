Convertir columnas de string a datetime.

```python
df_sin_fechas["FECHA_NACIMIENTO"].value_counts()
```

```python
df_sin_fechas["FECHA_NACIMIENTO"].dtype
```

```python
s = df_sin_fechas["FECHA_NACIMIENTO"].astype("string").str.strip()
s = s.str.replace(".", "/", regex=False).str.replace("-", "/", regex=False)

df_sin_fechas["FECHA_NACIMIENTO"] = pd.to_datetime(
    s,
    errors="coerce"
)

df_sin_fechas["FECHA_NACIMIENTO"].dtype
```

```python
df_sin_fechas.dtypes
```

```python
s = df_sin_fechas["FECHA_INGRESO"].astype("string").str.strip()
s = s.str.replace(".", "/", regex=False).str.replace("-", "/", regex=False)

df_sin_fechas["FECHA_INGRESO"] = pd.to_datetime(
    s,
    errors="coerce"
)

df_sin_fechas["FECHA_INGRESO"].dtype
```
